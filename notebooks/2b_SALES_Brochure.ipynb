```python
import os
import json
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import display, Markdown
from scraper import summarize_site, get_links 
```

```python
load_dotenv(override = True)
api_key = os.getenv("MY_OPENAI_API_KEY")
openai = OpenAI(api_key = api_key)
```

```python
def create_brochure(company_name, url):
    # Step 1: Gather the raw materials
    contents = summarize_site(url)
    links_found = get_links(url) 
    
    # Step 2: The "Marketing Strategist" Personality
    brochure_sys_prompt = f"""
    You are a Senior Conversion Copywriter. Your goal is to turn raw data into a 
    high-converting sales brochure following a specific marketing framework.

    MARKETING FRAMEWORK:
    1. ATTENTION-GRABBING HEADLINE: Start with a benefit-loaded headline that piques curiosity.
    2. BENEFIT-FOCUSED COPY: Transform features into outcomes (e.g., instead of "Fast CPU," use "Save 2 hours every day").
    3. UNIQUE SELLING PROPOSITION (USP): Clearly state what makes this company different.
    4. SOCIAL PROOF: Highlight client names, testimonials, or scale of impact if found in data.
    5. CLEAR CTA & CONTACT: Tell the user exactly what to do next.

    CRITICAL RULES:
    - NO HALLUCINATION: If social proof or a USP isn't in the data, do not invent it. 
    - VISUALS: Use '---' for section breaks and italics for sub-headlines.
    - Format in Markdown. No code blocks.
    """

    # Step 3: The Data Package with your checklist as the Template
    brochure_user_prompt = f"""
    COMPANY: {company_name}
    DATA SAMPLES:
    HOMEPAGE CONTENT: {contents}
    LINKS: {links_found}

    USE THIS STRUCTURE BASED ON PROFESSIONAL SALES STANDARDS:

    # [Benefit-Loaded Headline]
    *A sub-headline that hooks the reader's curiosity*
    ---
    ## The [Company Name] Difference
    *Our Unique Selling Proposition*
    (Write 1-2 paragraphs focusing on outcomes, not just features)

    ---
    ## Why Industry Leaders Trust Us
    *Social Proof & Impact*
    (List notable clients, stats, or culture highlights found in the data. If none exist, state the {company_name} scale of impact.)

    ---
    ## Solutions Designed for You
    *Core benefits and services*
    (List the top 3 benefits based on the HOMEPAGE CONTENT)

    ---
    ## Take the Next Step
    *Clear Call to Action*
    (Provide the most relevant contact/about links from the LINKS sample)

    VERIFICATION: Ensure every claim is grounded in the DATA SAMPLES or compan's real time information/update/news. Do not lie.
    """

    # Step 4: The OpenAI Call
    brochure_messages = [
        {"role": "system", "content": brochure_sys_prompt},
        {"role": "user", "content": brochure_user_prompt}
    ]

    response = openai.chat.completions.create(
        model="gpt-4o-mini", 
        messages=brochure_messages
    )

    result = response.choices[0].message.content
    
    # Step 5: Professional Display
    display(Markdown(result))
    return result
```

```python
create_brochure("HUGGING-FACE", "https://huggingface.co")
```
